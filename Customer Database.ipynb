```python
# Sales Amount to Product ID (Bar Graph)
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Read CSV file 
df = pd.read_csv('customer_master.csv')

# Count occurrences of each ProductID using Counter
product_counter = Counter(df['Product ID'])

# Sum sales amount for each ProductID
sales_sum = df.groupby('Product ID')['Sales Amount'].sum()

# Prepare data for plotting
products = list(sales_sum.index)
sales_values = list(sales_sum.values)

# Plot dashboard
plt.figure(figsize=(10, 6))
plt.bar(products, sales_values, color='skyblue')
plt.title('Total Sales Amount by Product ID', fontsize=16)
plt.xlabel('Product ID', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)
plt.xticks(rotation=45) # To rotate the x-axis characters by 45 degrees
plt.grid(axis='y', linestyle='--', alpha=0.7) # Grid lines along y- axis
plt.tight_layout()

# Show plot
plt.show()

# Print top 5 selling products by frequency
print("Top 5 Products by Number of Sales:")
for product, count in product_counter.most_common(5):
    print(f"Product {product}: {count} sales")
```

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

plt.style.use("fivethirtyeight")

# Read CSV file 
df = pd.read_csv('customer_master.csv')

# Simulating the Months
df['Month'] = np.tile(range(1,13), len(df) // 12 + 1)[:len(df)]

# Sum sales amount for each ProductID
sales = df.groupby(['Month', 'Customer Segment'])['Sales Amount'].sum().unstack(fill_value = 0)


# Plot dashboard for Stack Plot
plt.stackplot(sales.index, sales.T, labels = sales.columns, colors = ['#66b3ff','#99ff99','#ff9999'])

plt.legend(loc = 'upper left')

plt.title("Total Sales Amount by Customer Segment")
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.tight_layout()
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Read CSV file 
df = pd.read_csv('customer_master.csv')

# Count occurrences of each ProductID using Counter
segment_counter = Counter(df['Customer Segment'])

# Sum sales amount for each ProductID
sales_sum = df.groupby('Customer Segment')['Sales Amount'].sum()

# Prepare data for plotting
segments = list(sales_sum.index)
sales_values = list(sales_sum.values)

# Create dashboard figure 
plt.figure(figsize=(12,6))

# Plot dashboard for Bar Graph
plt.subplot(1,3,1)
plt.bar(segments, sales_values, color='skyblue')
plt.title('Total Sales Amount by Customer Segment', fontsize=16)
plt.xlabel('Customer Segment', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Plot dashboard for Pie Chart
plt.subplot(1,3,2)
plt.pie(sales_values, labels = segments, colors = ['#66b3ff','#99ff99','#ff9999'], wedgeprops = {'edgecolor' : 'black'})
plt.title('Total Sales Amount by Customer Segment', fontsize=16)

# Plot dashboard for Stack Plot
# Simulating the Months
df['Month'] = np.tile(range(1,13), len(df) // 12 + 1)[:len(df)]

# Sum sales amount for each ProductID
sales = df.groupby(['Month', 'Customer Segment'])['Sales Amount'].sum().unstack(fill_value = 0)

# Plot dashboard for Stack Plot
plt.subplot(1,3,3)
plt.stackplot(sales.index, sales.T, labels = sales.columns, colors = ['#66b3ff','#99ff99','#ff9999'])
plt.legend(loc = 'upper left')
plt.title("Total Sales Amount by Customer Segment")
plt.xlabel('Month')
plt.ylabel('Sales Amount')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv('customer_master.csv')

# Group by Customer ID and sum sales per customer
customer_sales = df.groupby('Customer ID')['Sales Amount'].sum()

# Calculate average sales
average_sales = customer_sales.mean()

# Create a bar chart showing each customer's sales
plt.figure(figsize=(10, 6))
plt.bar(customer_sales.index, customer_sales.values, color='skyblue')

# Add horizontal line for average sales
plt.axhline(average_sales, color='red', linestyle='--', label=f'Average Sales: {average_sales:.2f}')

# Labels and title
plt.xlabel('Customer ID')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=90, fontsize=4)
plt.title('Average Sales per Customer')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

# Print the average sales value
print(f"Average Sales per Customer: {average_sales:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('customer_master.csv')

# SALES BY COUNTRY
sales_by_country = df.groupby('Country')['Sales Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sales_by_country.plot(kind='bar', color='#66b3ff')
plt.title('Sales by Country',fontsize=16)
plt.ylabel('Total Sales Amount',fontsize=14)
plt.xlabel('Country',fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# SALES BY STATE
sales_by_state = df.groupby('State')['Sales Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sales_by_state.plot(kind='bar', color='#99ff99')
plt.title('Sales by State',fontsize=16)
plt.ylabel('Total Sales Amount',fontsize=14)
plt.xlabel('State',fontsize=14)
plt.xticks(rotation=90, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# SALES BY CITY
sales_by_city = df.groupby('City')['Sales Amount'].sum().sort_values(ascending=False).head(10)  # top 10 cities

plt.figure(figsize=(10, 6))
sales_by_city.plot(kind='bar', color='#ff9999')
plt.title('Top 10 Cities by Sales',fontsize=16)
plt.ylabel('Total Sales Amount',fontsize=14)
plt.xlabel('City',fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv('customer_master.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Remove non-numeric characters from License Cost (like $, AED, commas)
df['License Cost'] = df['License Cost'].replace('[^\d.]', '', regex=True)

# Convert to numeric
df['License Cost'] = pd.to_numeric(df['License Cost'], errors='coerce')

# Drop rows with missing Product ID or License Cost
df = df.dropna(subset=['Product ID', 'License Cost'])

# Group by Product ID and sum
product_license_cost = df.groupby('Product ID')['License Cost'].sum()

# Calculate average
average_license_cost = product_license_cost.mean()

# Plot
plt.figure(figsize=(10, 6))
plt.bar(product_license_cost.index.astype(str), product_license_cost.values, color='skyblue')
plt.axhline(average_license_cost, color='red', linestyle='--',
            label=f'Average License Cost: {average_license_cost:.2f}')
plt.xlabel('Product ID')
plt.ylabel('Total License Cost')
plt.xticks(rotation=90, fontsize=6)
plt.title('License Cost by Product')
plt.legend()
plt.tight_layout()
plt.show()

# Print average
print(f"Average License Cost per Product: {average_license_cost:.2f}")
```
